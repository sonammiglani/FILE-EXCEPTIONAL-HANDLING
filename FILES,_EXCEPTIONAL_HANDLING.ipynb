# Q12-What are the basic steps involved in exception handling in python?

A12-Exception handling in Python involves a structured approach to gracefully manage runtime errors. The basic steps are: Try block: Enclose the code that might raise an exception within a try block. Optional finally block: Regardless of whether an exception occurred or was caught, the code within the optional finally block will always execute. Raise statement: To explicitly raise an exception, the raise statement is used, followed by the exception to be raised. This is useful for custom error handling or re-raising exceptions

# Q15-How does pythons's garbage collection system works?

Python employs a hybrid garbage collection system that combines reference counting with a generational garbage collector to manage memory automatically, identifying and reclaiming memory occupied by unused objects. Here's a breakdown: Reference Counting: Python keeps track of how many references (pointers) point to each object. When an object's reference count reaches zero, meaning no part of the program is using it, the object is eligible for garbage collection and its memory is freed. Generational Garbage Collection: This mechanism addresses the limitations of pure reference counting, particularly with circular references (where objects reference each other and their reference counts never reach zero). Python classifies objects into generations (0, 1, and 2) based on how long they've survived previous garbage collection cycles. New objects are placed in generation 0, and if they survive a collection, they move to generation 1, and so on. The garbage collector runs periodically, focusing on younger generations first, and less frequently on older generations. The generational garbage collector uses a "mark and sweep" algorithm to identify and collect unreachable objects, including those involved in circular references. Automatic Memory Management: The garbage collector operates automatically in the background, so developers don't need to manually allocate or deallocate memory. gc module: Python provides a gc module that allows developers to access and control the garbage collector, though it's generally not necessary to use it directly in most applications.

# Q18-What is the differnce between os.fork() and multiprocessing in python?

A18-In Python, os.fork() is a lower-level, OS-specific mechanism to create a new process, while multiprocessing provides a higher-level, portable framework for managing multiple processes and leveraging multiple CPU cores. os.fork() is limited to Unix-like systems and directly uses the operating system's fork call, creating a nearly identical copy of the parent process. multiprocessing, on the other hand, offers different start methods (including fork, spawn, and forkserver) and provides tools for inter-process communication and synchronization, making it more versatile and easier to use across different platforms.

```python
#Q12- Write a program that demonstrate using multiple except blocks to handle different types of exception.
# Function to demonstrate handling multiple exceptions
def handle_exceptions():
    try:
        # Taking input from the user
        x = int(input("Enter a number: "))
        y = int(input("Enter another number: "))

        # Perform division
        result = x / y
        print(f"The result of {x} divided by {y} is: {result}")

        # Attempt to access an index that may be out of range
        my_list = [1, 2, 3]
        print(f"The 5th element of the list is: {my_list[4]}")

    except ZeroDivisionError:
        print("Error: You can't divide by zero!")
    except ValueError:
        print("Error: Invalid input! Please enter a valid number.")
    except IndexError:
        print("Error: List index out of range!")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Call the function
handle_exceptions()
```

```python
#Q18-How would you implement a basic logging setup that logs to a file with rotation after 1 MB
import logging
from logging.handlers import RotatingFileHandler

# Create a rotating file handler
log_handler = RotatingFileHandler(
    'app.log',       # Log file name
    maxBytes=1 * 1024 * 1024,  # 1MB size limit
    backupCount=5             # Keep 5 backup logs
)

# Set format for log messages
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
log_handler.setFormatter(formatter)

# Set up the logger
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)       # Log level
logger.addHandler(log_handler)

# Example log entries
logger.info("This is an info message.")
logger.warning("This is a warning.")
logger.error("This is an error message.")
```

```python
#Q21- Write a python program that reads a file and prints the number  of occurences of a specific word
def count_word_occurrences(file_name, word_to_count):
    try:
        with open(file_name, "r") as file:
            content = file.read()

            # Convert both content and word to lowercase for case-insensitive search
            word_count = content.lower().split().count(word_to_count.lower())

            print(f"The word '{word_to_count}' appears {word_count} time(s) in the file.")
    except FileNotFoundError:
        print(f"Error: File '{file_name}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file_name = "example.txt"         # Replace with your file name
word_to_count = "python"          # Replace with the word you're searching for
count_word_occurrences(file_name, word_to_count)
```

```python
#Q22-How can you check if a file is empty before attempting to read its content?
import os

def check_if_empty(file_name):
    # Check if the file exists and is not empty
    if os.path.exists(file_name) and os.stat(file_name).st_size > 0:
        with open(file_name, "r") as file:
            content = file.read()
            print("File contents:\n", content)
    else:
        print(f"The file '{file_name}' is empty or doesn't exist.")

# Example usage
file_name = "example.txt"
check_if_empty(file_name)
```
